```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
import pandas as pd
```

```python
# Loads in fort.14 31k mesh
mesh_data_31k = loadmat('3D_mesh_plot.mat')
mesh_31k      = mesh_data_31k['S']
lat31         = np.array(mesh_31k[:,0]) # latitude 31k
lon31         = np.array(mesh_31k[:,1]) # longitude 31k
bath31        = mesh_31k[3,:] # bathymetry 31k
```

```python
# Loads in fort.14 8k mesh
mesh_data_8k = loadmat('fort14_8k.mat')
mesh_8k      = mesh_data_8k['mesh']
lat8         = np.array(mesh_8k[:, 0]) # latitude 8k
lon8         = np.array(mesh_8k[:, 1]) # longitude 8k

```
